```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister

# Define the quantum register
qr = QuantumRegister(8)
cr= ClassicalRegister(8)
# Define the quantum circuit
qc = QuantumCircuit(qr,cr)

# Prepare each qubit in the state |+>
for i in range(8):
    qc.h(qr[i])
```

```python
print(qc)
```

```python
import random
# Define the basis choice (X or Y) for 8 qubits
basis_choice_alice = [random.choice(['X', 'Y']) for _ in range(8)]
basis_choice_bob = [random.choice(['X', 'Y']) for _ in range(8)]

# Keep the basis choice secret (just print for demonstration purposes)
print("Alice's basis choice:", basis_choice_alice)
print("Bob's basis choice:", basis_choice_bob)
```

```python
# Define the angle of rotation (e.g., π/4)
theta = 3.14159 / 4

# Encode the qubit based on Alice's basis choice
for i in range(8):
  if basis_choice_alice[i] == 'X':
    qc.rx(theta, qr[i])
  elif basis_choice_alice[i] == 'Y':
    qc.ry(theta, qr[i])
```

```python
# Measure the qubit in the chosen basis (X or Y)
if basis_choice_bob == 'X':
    measurement_bob=qc.measure(qr[0], cr[0])
elif basis_choice_bob == 'Y':
    qc.ry(-3.14159/2, qr[0])
    measurement_bob=qc.measure(qr[0], cr[0])


# Measure Alice's original qubit in her chosen basis (X or Y)
if basis_choice_alice == 'X':
    qc.measure(qr[0], cr[0])
elif basis_choice_alice == 'Y':
    qc.ry(-3.14159/2, qr[0])
    qc.measure(qr[0], cr[0])
```

```python
def classical_post_processing(basis_choice_alice, basis_choice_bob, measurement_alice, measurement_bob):
    shared_key = ''

    # Compare basis choices and keep or discard measurement outcomes
    for i in range(len(basis_choice_alice)):
        if basis_choice_alice[i] == basis_choice_bob[i]:
            shared_key += measurement_alice[i]
        else:
            continue

    return shared_key

# Repeat steps 1-6 multiple times to generate a shared key
shared_key = ''
for _ in range(1000):  # Repeat 1000 times
    # Steps 1-5 code here...
    shared_key += classical_post_processing(basis_choice_alice, basis_choice_bob, measurement_alice, measurement_bob)

print('Shared key:', shared_key)
```
