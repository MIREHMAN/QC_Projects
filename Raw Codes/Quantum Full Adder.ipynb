```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister,


class FullAdderCircuit:
    def __init__(self):
        self.a = QuantumRegister(1, 'a')  # First input bit
        self.b = QuantumRegister(1, 'b')  # Second input bit
        self.cin = QuantumRegister(1, 'cin')  # Carry input bit
        self.sum_out = QuantumRegister(1, 'sum_out')  # Sum output bit
        self.cout = QuantumRegister(1, 'cout')  # Carry output bit
        self.c = ClassicalRegister(2, 'c')  # Classical register for measurement
        self.qc = QuantumCircuit(self.a, self.b, self.cin, self.sum_out, self.cout, self.c)

        # Define the quantum circuits for individual components
        self.half_adder = QuantumCircuit(3, name="Half_Adder")
        self.half_adder.cx(0, 1)
        self.half_adder.cx(0, 2)
        self.half_adder.ccx(1, 2, 0)  # Compute the sum
        self.half_adder.to_instruction()

        self.full_adder = QuantumCircuit(5, name="Full_Adder")
        self.full_adder.ccx(0, 1, 3)  # Compute the carry
        self.full_adder.cx(0, 1)  # Propagate the carry
        self.full_adder.cx(1, 2)  # Propagate the carry
        self.full_adder.cx(3, 4)  # Propagate the carry
        self.full_adder.to_instruction()

    def run_circuit(self, input_values):
        self.qc.data = []  # Clear previous circuit data

        a_val, b_val, cin_val = input_values

        # Setting input values
        if a_val == '1':
            self.qc.x(self.a[0])
        if b_val == '1':
            self.qc.x(self.b[0])
        if cin_val == '1':
            self.qc.x(self.cin[0])

        self.qc.append(self.half_adder, [self.a[0], self.b[0], self.sum_out[0]])
        self.qc.append(self.full_adder, [self.a[0], self.b[0], self.cin[0], self.sum_out[0], self.cout[0]])

        # Measure output bits
        self.qc.measure(self.sum_out[0], self.c[0])
        self.qc.measure(self.cout[0], self.c[1])

        # Execute the circuit
        simulator = Aer.get_backend('qasm_simulator')
        job = execute(self.qc, simulator, shots=1024)
        result = job.result()
        counts = result.get_counts(self.qc)
        print(f"For inputs a={a_val}, b={b_val}, cin={cin_val}, Total counts are:", counts)

# Example usage:
adder_circuit = FullAdderCircuit()
inputs = ['000', '001', '010', '011', '100', '101', '110', '111']

for input_values in inputs:
    adder_circuit.run_circuit(input_values)
```
