```python
import matplotlib.pyplot as plt
import numpy as np


# Data
models = ['Classical SVM', 'Classical RF', 'Quantum VQC', 'Quantum SVC Corr Kernel', 'Quantum SVC PCA Kernel']
accuracy = [0.8943, 0.9736, 0.98, 0.94, 0.89]
precision = [0.8943, 0.9752, 0.92, 0.9663, 0.9418]
recall = [1.000, 0.9958, 0.96, 0.9663, 0.89]
f1_score = [0.9442, 0.9854, 1.00, 0.9663, 1.00]


# Line Plot
plt.figure(figsize=(10, 6))
plt.plot(models, accuracy, label='Accuracy', marker='o', linestyle='-', color='blue')
plt.plot(models, precision, label='Precision', marker='s', linestyle='--', color='green')
plt.plot(models, recall, label='Recall', marker='^', linestyle='-.', color='red')
plt.plot(models, f1_score, label='F1 Score', marker='*', linestyle=':', color='purple')
plt.title('African Crisis - Metrics Comparison')
plt.xlabel('Model')
plt.ylabel('Metric Value')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Data
models = ['Classical SVM', 'Classical RF', 'Quantum VQC', 'Quantum SVC Corr Kernel', 'Quantum SVC PCA Kernel']
accuracy = [0.8943, 0.9736, 0.98, 0.94, 0.89]
precision = [0.8943, 0.9752, 0.92, 0.9663, 0.9418]
recall = [1.000, 0.9958, 0.96, 0.9663, 0.89]
f1_score = [0.9442, 0.9854, 1.00, 0.9663, 1.00]


# Grouped Bar Graph
x = np.arange(len(models))
width = 0.2


fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - 1.5*width, accuracy, width, label='Accuracy', color='blue')
rects2 = ax.bar(x - 0.5*width, precision, width, label='Precision', color='green')
rects3 = ax.bar(x + 0.5*width, recall, width, label='Recall', color='red')
rects4 = ax.bar(x + 1.5*width, f1_score, width, label='F1 Score', color='purple')


ax.set_title('African Crisis - Metrics Comparison')
ax.set_xlabel('Model')
ax.set_ylabel('Metric Value')
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=45)
ax.legend(loc='upper right')
plt.show()
```
